```python
# imported required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loaded the data set
df=pd.read_csv(r"c:\Users\DELL\Downloads\Dataset .csv")
```

```python
#LEVEL 1
#TASK 1
#Task: Top Cuisines
```

```python
#Determine the top three most common cuisines in the dataset
top_3_cuisine=df['Cuisines'].str.split(', ').explode('Cuisines').value_counts().head(3)
top_3_cuisine
```

```python
plt.figure(figsize=(7,7))
plt.bar(top_3_cuisine.index,top_3_cuisine.values,color=['green'])
plt.xlabel('Cuisine')
plt.ylabel('no of occurences')
plt.show()
```

```python
#Calculate the percentage of restaurants that serve each of the top cuisines.
top_cuisines=df['Cuisines'].str.split(', ').explode('Cuisines').value_counts().head()
```

```python
total_no_restaurants=df.shape[0]
```

```python

top_cuisine_percentage=(top_cuisines/total_no_restaurants)*100
top_cuisine_percentage
```

```python
plt.figure(figsize=(7,7))
plt.bar(top_cuisine_percentage.index,top_cuisine_percentage.values,color='g')
plt.xlabel('Top_cuisines')
plt.ylabel("restaurant percentage")
plt.show()
```
