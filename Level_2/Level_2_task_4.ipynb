```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loaded the data set
df=pd.read_csv(r"c:\Users\DELL\Downloads\Dataset .csv")
```

```python
df.columns
```

```python
# Identify if there are any restaurant chains present in the dataset.
restaurant_chains=df[['Restaurant Name','Address']].groupby('Restaurant Name').nunique()
chains = restaurant_chains[restaurant_chains > 1]
top_chains=chains.sort_values(by="Address",ascending=False).head(10)
top_chains
```

```python
# Visualizing Top 10 Restaurants by location
plt.figure(figsize=(10,6))
plt.bar(top_chains.index,top_chains["Address"],color='yellow')
plt.xlabel("Restaurant Name",fontsize=12)
plt.ylabel("No of unique locations")
plt.title("Top 10 Restaurants by location",fontsize=14)
plt.show()
```

```python
# Analyze the ratings and popularity of different restaurant chains.
df.columns
```

```python
chain_analysis=df[['Restaurant Name','Aggregate rating','Votes']].groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum'
}).reset_index()
```

```python
top_rated_chains=chain_analysis.sort_values(by='Aggregate rating',ascending=False).head(10)
most_popular_chains=chain_analysis.sort_values(by='Votes',ascending=False).head(10)
```
