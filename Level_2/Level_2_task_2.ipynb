```python
# imported required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loaded data set
df=pd.read_csv(r"c:\Users\DELL\Downloads\Dataset .csv")
```

```python
#Identify the most common combinations of cuisines in the dataset
```

```python
# checking the null values in cuisines column
df[df['Cuisines'].isnull()]
```

```python
# dropping null values in the cuisines column
df.dropna(subset='Cuisines')
```

```python
cuisines_count=df['Cuisines'].value_counts(ascending=False)
```

```python
# Top most common combination of cuisines
most_common_cuisines=cuisines_count[cuisines_count.index.str.contains(',')].head()
most_common_cuisines
```

```python
# piechart visualization for most common cuisines
plt.figure(figsize=(4,5))
plt.pie(most_common_cuisines.values,labels=most_common_cuisines.index,autopct="%.2f%%")
plt.title("most common cuisines")
plt.show()
```
